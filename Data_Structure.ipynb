Data Types and Structures Assignment:-


1. What are data structures, and why are they important?

   - Data structures are specialized formats used to store, organize, and manage data efficiently. They play a crucial role in optimizing performance, particularly in searching, sorting, and data manipulation tasks.

2. Explain the difference between mutable and immutable data types with examples?

    - Mutable data types allow modifications after creation, such as lists (list in Python), whereas immutable data types, like tuples (tuple) and strings (str), cannot be altered once created. For instance, modifying a list [1,2,3] is possible, but a tuple (1,2,3) remains unchanged.

3. What are the main differences between lists and tuples in Python?

    - Lists are dynamic and can be modified, whereas tuples are immutable and consume less memory.
    Lists are generally used when the data needs to be changed frequently, while tuples are ideal for fixed values.

4. Describe how dictionaries store data?

    - Dictionaries store data as key-value pairs, utilizing a hash function to quickly retrieve values based on their unique keys.

5. Why might you use a set instead of a list in Python?

    - A set is useful when uniqueness is required, as it automatically removes duplicate values and provides faster membership checks than a list.

6. What is a string in Python, and how is it different from a list?

    - A string is an immutable sequence of characters, while a list is a mutable collection of elements that can hold different data types.

7. How do tuples ensure data integrity in Python?

    - Tuples are immutable, meaning their values cannot be changed after creation, which ensures that important data remains unchanged throughout program execution.

8. What is a hash table, and how does it relate to dictionaries in Python?

    - A hash table is a data structure that stores key-value pairs using a hashing mechanism. Python dictionaries are built on hash tables, allowing fast data retrieval.

9. Can lists contain different data types in Python?

    - Yes, Python lists can store elements of multiple data types, including integers, strings, and even other lists.

10. Explain why strings are immutable in Python?

    - Strings are immutable to improve memory efficiency, enable safe string interning, and prevent accidental modifications.

11. What advantages do dictionaries offer over lists for certain tasks?

    - Dictionaries provide O(1) time complexity for key-based lookups, whereas lists require O(n) time complexity for searching elements.

12. Describe a scenario where using a tuple would be preferable over a list?

    - When storing fixed configuration settings or constant values that should not be changed, tuples are preferable over lists.

13. How do sets handle duplicate values in Python?

    - Sets automatically remove duplicate elements, ensuring that only unique values are stored.

14. How does the “in” keyword work differently for lists and dictionaries?

    - In lists, in checks for the presence of a value, whereas in dictionaries, it checks for the presence of a key.

15. Can you modify the elements of a tuple? Explain why or why not.

    - No, tuples are immutable, meaning their elements cannot be altered once assigned.

16. What is a nested dictionary, and give an example of its use case?

    - A nested dictionary is a dictionary containing another dictionary as a value.
    Example: student = {"name": "John", "grades": {"math": 90, "science": 85}}

17. Describe the time complexity of accessing elements in a dictionary?

    - Dictionary lookups have an average time complexity of O(1) due to their underlying hash table implementation.

18. In what situations are lists preferred over dictionaries?

    - Lists are preferable when elements need to be stored sequentially and accessed via index positions rather than unique keys.

19. Why are dictionaries considered unordered, and how does that affect data
retrieval?

    - Prior to Python 3.7, dictionaries did not maintain insertion order. However, in newer versions, while they do maintain order, they are still optimized for key-based retrieval rather than sequential access.

20. Explain the difference between a list and a dictionary in terms of data retrieval?
